```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset from a CSV file
df = pd.read_csv('cancer programs.csv')

# Create a scatter plot for 'Cases' vs 'Project'
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Cases', y='Project')
plt.title('Cases vs Project')
plt.xlabel('Cases')
plt.ylabel('Project')
plt.grid(True)
plt.savefig('scatter_plot.png')  # Save the scatter plot as 'scatter_plot.png'
plt.show()

# Create a bar plot for 'Project' vs 'Disease Type'
plt.figure(figsize=(12, 8))
sns.countplot(data=df, y='Project', hue='Disease Type', palette='Set2')
plt.title('Distribution of Disease Types for Each Project')
plt.xlabel('Count')
plt.ylabel('Project')
plt.legend(title='Disease Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.savefig('project_vs_disease_type_bar_plot.png')  # Save the bar plot as 'project_vs_disease_type_bar_plot.png'
plt.show()

# Create a pie chart for 'Experimental Strategy'
plt.figure(figsize=(8, 8))
df['Experimental Strategy'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Experimental Strategy Distribution')
plt.ylabel('')
plt.savefig('pie_chart.png')  # Save the pie chart as 'pie_chart.png'
plt.show()
```
